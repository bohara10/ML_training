```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
df.head(5)
```

```python
df["Date"] = df["Date"].replace({"'":""},regex=True)
df.head(5)
```

```python
# @title Maxpulse

from matplotlib import pyplot as plt
df['Maxpulse'].plot(kind='hist', bins=20, title='Maxpulse')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df
```

```python
#return a dataset with no empty cells
#df.dropna() will remove empty cells but will not change the dataframe
#if we want date frame to be changed , use inplace=true

new_df = df.dropna()
new_df
```

```python
# another way to handling the null value is by replacing the null value with 130
df_new2 = df.fillna(130)
df_new2
```

```python
# fill replace
df['Calories'].fillna(150, inplace=True)
df
```

```python
# fill null values with mean , median or mode

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
x= df["Calories"].mean()
df["Calories"].fillna(x,inplace = True)
df
```
