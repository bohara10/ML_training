# Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basic of matplotlib

#line plot

x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="Sine wave")
plt.title("Line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend() #used to indicate the line with specific color
plt.show()
```

```python
# Bar chart

categories = ['A', 'B', 'C']
values = [10, 20, 30]
plt.bar(categories, values, color =['red', 'blue', 'green'])
plt.title("Bar chart example")
plt.xlabel("categories")
plt.ylabel("values")

plt.show()
```

# Seaborn


```python
import seaborn as sns
# Histogram and KDE plot:
data = np.random.randn(1000)
sns.histplot(data, kde=True, color= 'red')
plt.title("Histogram with KDE")
plt.show()
```

```python
# heatmap
matrix = np.random.rand(5,5)
sns.heatmap(matrix, annot= True, cmap="RdBu")
plt.title("Headmap example ")
plt.show()
```

```python
#pairplot example

sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "d"])
plt.show()
```
